```python
import os, shutil, random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam


# ===================== SETUP =====================
IMG_SIZE = 224
original_data_dir = '/content/drive/MyDrive/data1'
base_dir = '/content/data1_split'
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# ===================== SPLIT DATA =====================
classes = [folder for folder in os.listdir(original_data_dir) if os.path.isdir(os.path.join(original_data_dir, folder))]

if os.path.exists(base_dir):
    shutil.rmtree(base_dir)

os.makedirs(train_dir)
os.makedirs(test_dir)

for cls in classes:
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

for cls in classes:
    folder_path = os.path.join(original_data_dir, cls)
    images = [img for img in os.listdir(folder_path) if img.lower().endswith(('.jpg', '.png', '.jpeg'))]
    random.shuffle(images)
    split_idx = int(0.8 * len(images))
    train_imgs = images[:split_idx]
    test_imgs = images[split_idx:]

    for img in train_imgs:
        shutil.copy(os.path.join(folder_path, img), os.path.join(train_dir, cls, img))
    for img in test_imgs:
        shutil.copy(os.path.join(folder_path, img), os.path.join(test_dir, cls, img))

# ===================== DATA AUGMENTATION =====================
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=32, class_mode='categorical', shuffle=True
)

val_data = val_datagen.flow_from_directory(
    test_dir, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=32, class_mode='categorical', shuffle=False
)
```

```python
print("\nüß† Training CNN from Scratch...")

cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(len(classes), activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.fit(train_data, validation_data=val_data, epochs=30)
```

```python
print("\nüìä Evaluating CNN Model...")

# Get class labels
class_labels = list(train_data.class_indices.keys())

# Predict on validation data
val_preds = cnn_model.predict(val_data)
y_pred = np.argmax(val_preds, axis=1)
y_true = val_data.classes

# Classification Report (Fixed)
labels = list(range(len(class_labels)))  # Ensure all class indices are included
print("üßæ Classification Report:")
print(classification_report(y_true, y_pred, labels=labels, target_names=class_labels))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.title("üß© CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
0# ===================== 6. BUILD & TRAIN RESNET50 MODEL# ===================== 6. BUILD & TRAIN RESNET50 MODEL =====================
print("\nüîÅ Training ResNet50 Transfer Learning...")

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.2)(x)
output = Dense(len(classes), activation='softmax')(x)

resnet_model = Model(inputs=base_model.input, outputs=output)
resnet_model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

resnet_model.fit(train_data, validation_data=val_data, epochs=70)  # Change epochs as needed

# ‚úÖ Fine-tuning (Unfreeze last 30 layers)
for layer in base_model.layers[-30:]:
    layer.trainable = True

resnet_model.compile(optimizer=Adam(1e-5), loss='categorical_crossentropy', metrics=['accuracy'])
resnet_model.fit(train_data, validation_data=val_data, epochs=80)
```

```python
print("\nüìä Evaluating ResNet50...")
resnet_preds = resnet_model.predict(val_data)
resnet_y_pred = np.argmax(resnet_preds, axis=1)

print("\nResNet50 Classification Report:")
print(classification_report(y_true, resnet_y_pred, labels = labels,target_names=classes))

cm = confusion_matrix(y_true, resnet_y_pred)
sns.heatmap(cm, annot=True, xticklabels=classes, yticklabels=classes, fmt='d', cmap='Greens')
plt.title("ResNet50 Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
import random
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
import os

def predict_single_image(model, image_path, classes):
    img = load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    predicted_class = classes[np.argmax(prediction)]
    confidence = np.max(prediction) * 100

    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {predicted_class} ({confidence:.2f}%)")
    plt.show()

def predict_random_image_from_test(model, test_dir, classes):
    class_choice = random.choice(classes)
    class_path = os.path.join(test_dir, class_choice)
    image_file = random.choice(os.listdir(class_path))
    image_path = os.path.join(class_path, image_file)

    print(f"\nüîç Predicting random image: {image_path}")
    predict_single_image(model, image_path, classes)

# ‚úÖ Example usage
predict_random_image_from_test(resnet_model, test_dir, classes)
```

```python
import pandas as pd

# Assuming:
# - report_resnet50 is the classification report from ResNet50 (transfer learning)
# - report_cnn is the classification report from your CNN (built from scratch)

# Get all unique class labels excluding "accuracy"
labels = sorted(set(report_resnet50.keys()).union(set(report_cnn.keys())) - {'accuracy'})

# Initialize table
comparison = {
    "Class": [],
    "Precision (CNN)": [],
    "Precision (ResNet50)": [],
    "Recall (CNN)": [],
    "Recall (ResNet50)": [],
    "F1-Score (CNN)": [],
    "F1-Score (ResNet50)": []
}

for label in labels:
    # CNN values
    cnn_prec = report_cnn.get(label, {}).get("precision", 0.0)
    cnn_rec = report_cnn.get(label, {}).get("recall", 0.0)
    cnn_f1 = report_cnn.get(label, {}).get("f1-score", 0.0)

    # ResNet50 values
    r50_prec = report_resnet50.get(label, {}).get("precision", 0.0)
    r50_rec = report_resnet50.get(label, {}).get("recall", 0.0)
    r50_f1 = report_resnet50.get(label, {}).get("f1-score", 0.0)

    # Append to comparison table
    comparison["Class"].append(label)
    comparison["Precision (CNN)"].append(round(cnn_prec, 2))
    comparison["Precision (ResNet50)"].append(round(r50_prec, 2))
    comparison["Recall (CNN)"].append(round(cnn_rec, 2))
    comparison["Recall (ResNet50)"].append(round(r50_rec, 2))
    comparison["F1-Score (CNN)"].append(round(cnn_f1, 2))
    comparison["F1-Score (ResNet50)"].append(round(r50_f1, 2))

# Add accuracy, macro avg, weighted avg (manually handle accuracy as it's not a dict)
for metric in ["accuracy", "macro avg", "weighted avg"]:
    if metric == "accuracy":
        cnn_acc = report_cnn.get("accuracy", 0.0)
        r50_acc = report_resnet50.get("accuracy", 0.0)
        comparison["Class"].append("accuracy")
        comparison["Precision (CNN)"].append(round(cnn_acc, 2))
        comparison["Precision (ResNet50)"].append(round(r50_acc, 2))
        comparison["Recall (CNN)"].append(round(cnn_acc, 2))
        comparison["Recall (ResNet50)"].append(round(r50_acc, 2))
        comparison["F1-Score (CNN)"].append(round(cnn_acc, 2))
        comparison["F1-Score (ResNet50)"].append(round(r50_acc, 2))
    else:
        comparison["Class"].append(metric)
        comparison["Precision (CNN)"].append(round(report_cnn[metric]["precision"], 2))
        comparison["Precision (ResNet50)"].append(round(report_resnet50[metric]["precision"], 2))
        comparison["Recall (CNN)"].append(round(report_cnn[metric]["recall"], 2))
        comparison["Recall (ResNet50)"].append(round(report_resnet50[metric]["recall"], 2))
        comparison["F1-Score (CNN)"].append(round(report_cnn[metric]["f1-score"], 2))
        comparison["F1-Score (ResNet50)"].append(round(report_resnet50[metric]["f1-score"], 2))

# Create DataFrame
cnn_vs_r50_df = pd.DataFrame(comparison)
cnn_vs_r50_df.set_index("Class", inplace=True)

# Show table
cnn_vs_r50_df
```
